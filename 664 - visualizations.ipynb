```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
import seaborn as sns
rcParams['figure.figsize'] = 11, 7
```

```python
df = pd.read_csv(r"C:/Users/RAHUL/Downloads/WA_Fn-UseC_-Marketing-Campaign-Plan_-GroceryS (4).csv")
```

```python
############1st parameter##############
new1 = df.groupby('HealthyStore',as_index=False).count()[['HealthyStore','vari']]
```

```python
new1.head()
```

```python
new1 = new1.sort_values('vari',ascending=False)
```

```python
new1
```

```python
mul = new1[new1.vari > 1]
```

```python
mul
```

```python
HealthyStore_1 = mul['HealthyStore'].values
no1 = mul['vari'].values
```

```python
rcParams['figure.figsize'] = 30, 15
sns.barplot(HealthyStore_1,no1).set_xticklabels(HealthyStore_1,rotation = 65)
plt.title("Total number of stores")
plt.xlabel("Type of stores")
plt.ylabel("total number of stores")
```

```python
########1st parameter vs amt#######
```

```python
########plotted vs amtspent#####
dum = df.groupby('HealthyStore',as_index=False).sum()

import matplotlib.pyplot as pls 
rcParams['figure.figsize'] = 27, 8
sns.barplot(dum.HealthyStore,dum.AmountSpent)
#df.plot(x='WhoShoppingFor', y='AmountSpent', kind='bar') 
plt.title("Total number of stores vs Amount spent")
plt.show()
```
